```python
from deepface import DeepFace
from google.colab import files
import matplotlib.pyplot as plt
import cv2

# upload image
uploaded = files.upload()

# get file name
img_path = list(uploaded.keys())[0]

# show image
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis("off")
plt.show()

# run analysis
result = DeepFace.analyze(img_path, actions=['age', 'gender', 'emotion'], enforce_detection=False)

print(result)
```

```python
# extract results
age = result[0]['age']
gender = result[0]['dominant_gender']
emotion = result[0]['dominant_emotion']

# convert age to binary category
if age >= 50:
    age_group = "Elderly"
else:
    age_group = "Adult"

# convert emotion to happy/sad
if emotion == "happy":
    emotion_binary = "Happy"
else:
    emotion_binary = "Sad"

# print simplified results
print("Binary classification results:")
print("Age group:", age_group)
print("Gender:", gender)
print("Expression:", emotion_binary)
```

```python
import os
import time
from google.colab import files
from deepface import DeepFace
import pandas as pd

uploaded = files.upload()

results = []
times_ms = []

for img_path in uploaded.keys():

    # measure inference time (DeepFace analyze)
    t0 = time.perf_counter()
    analysis = DeepFace.analyze(
        img_path,
        actions=['age', 'gender', 'emotion'],
        enforce_detection=False
    )
    t1 = time.perf_counter()
    infer_ms = (t1 - t0) * 1000
    times_ms.append(infer_ms)

    age = analysis[0]['age']
    gender = str(analysis[0]['dominant_gender']).strip().lower()
    emotion = str(analysis[0]['dominant_emotion']).strip().lower()

    # age mapping (binary)
    age_group = "Elderly" if age >= 50 else "Adult"

    # gender mapping (binary normalize)
    if gender in ["woman", "female", "f"]:
        gender_binary = "Woman"
    else:
        gender_binary = "Man"

    # emotion mapping (binary)
    emotion_binary = "Happy" if emotion == "happy" else "Sad"

    # extract ground truth from filename
    filename = os.path.splitext(img_path)[0]
    parts = filename.split("_")

    # Expected filename format:
    # 01_adult_man_happy.jpg
    true_age = parts[1].capitalize()       # Adult/Elderly
    true_gender = parts[2].capitalize()    # Man/Woman
    true_emotion = parts[3].capitalize()   # Happy/Sad

    results.append({
        "image": img_path,
        "true_age": true_age,
        "pred_age": age_group,
        "true_gender": true_gender,
        "pred_gender": gender_binary,
        "true_emotion": true_emotion,
        "pred_emotion": emotion_binary,
        "inference_time_ms": round(infer_ms, 2)
    })

df = pd.DataFrame(results)

print("Average inference time (ms):", round(df["inference_time_ms"].mean(), 2))
df
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

tasks = {
    "Age": ("true_age", "pred_age"),
    "Gender": ("true_gender", "pred_gender"),
    "Emotion": ("true_emotion", "pred_emotion")
}

metrics = []

for task, (true_col, pred_col) in tasks.items():

    accuracy = accuracy_score(df[true_col], df[pred_col])
    precision = precision_score(df[true_col], df[pred_col], average='macro')
    recall = recall_score(df[true_col], df[pred_col], average='macro')
    f1 = f1_score(df[true_col], df[pred_col], average='macro')

    metrics.append({
        "Task": task,
        "Accuracy": round(accuracy,3),
        "Precision": round(precision,3),
        "Recall": round(recall,3),
        "F1 Score": round(f1,3)
    })

metrics_df = pd.DataFrame(metrics)

metrics_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def plot_cm(true_col, pred_col, title, filename):

    cm = confusion_matrix(df[true_col], df[pred_col])

    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=df[true_col].unique(),
                yticklabels=df[true_col].unique())

    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(title)

    plt.savefig(filename)
    plt.show()

plot_cm("true_age", "pred_age", "Confusion Matrix — Age", "cm_age_group.png")
plot_cm("true_gender", "pred_gender", "Confusion Matrix — Gender", "cm_gender.png")
plot_cm("true_emotion", "pred_emotion", "Confusion Matrix — Emotion", "cm_expression.png")
```

```python
df.to_csv("evaluation_results.csv", index=False)
metrics_df.to_csv("metrics_summary.csv", index=False)

files.download("evaluation_results.csv")
files.download("metrics_summary.csv")
```
